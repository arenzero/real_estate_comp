```python
#和暦 to 西暦
import re

def wareki_seireki(wareki_list):
    for i in range(len(wareki_list)):
        try:
            if re.match('昭和元年', wareki_list[i]):
                wareki_list[i] = 1926

            elif re.match('昭和',wareki_list[i]):
                wareki_list[i] = 1925 + int(re.sub(r'\D','',wareki_list[i]))

            elif re.match('平成元年', wareki_list[i]):
                wareki_list[i] = 1989

            elif re.match('平成',wareki_list[i]):
                wareki_list[i] = 1988 + int(re.sub(r'\D','',wareki_list[i]))
        except:
            continue
```

```python
train = pd.read_csv("../processed_data/merged_train.csv")
test = pd.read_csv("../processed_data/merged_test.csv")

print(train.shape,test.shape)

#用途FE用
youto = train["用途"].str.split("、",expand=True)
youto_master = pd.get_dummies(youto[0]).copy()

def FE(df):

    #建築年をFE
    df["建築年_元号"] = df["建築年"].replace({np.nan:0,"戦前":1,"(.*)昭和":2,"(.*)平成":3},regex=True)

    #最寄駅：距離をFE
    df['最寄駅：距離（分）'] = df["最寄駅：距離（分）"].replace({'30分?60分':'45','1H?1H30':'75','1H30?2H':"105",'2H?':"150"}).astype(float)

    #面積をFE
    df["面積（㎡）"] = df["面積（㎡）"].replace({'2000㎡以上':'3000','5000㎡以上':'6000'}).astype(float)

    #いらないのを削除
    df = df.drop(["市区町村名","都道府県名"],axis=1)

    #間取りを順序変数化
    madori = ['１Ｒ','１Ｒ＋Ｓ','１Ｋ','１Ｋ＋Ｓ','１ＤＫ','１ＤＫ＋Ｓ',
     '１Ｌ','１Ｌ＋Ｓ','１ＬＫ','１ＬＤ＋Ｓ','１ＬＤＫ','１ＬＤＫ＋Ｓ','２Ｋ','２Ｋ＋Ｓ','２Ｄ','２ＤＫ','２ＤＫ＋Ｓ',
     '２ＬＫ','２ＬＫ＋Ｓ','２ＬＤ','２ＬＤ＋Ｓ','２ＬＤＫ','２ＬＤＫ＋Ｓ','３Ｋ','３Ｋ＋Ｓ','３ＤＫ','３ＤＫ＋Ｓ',
     '３ＬＫ','３ＬＤ','３ＬＤ＋Ｓ','３ＬＤＫ','３ＬＤＫ＋Ｓ','３ＬＤＫ＋Ｋ','４Ｋ','４ＤＫ','４ＤＫ＋Ｓ',
     '４ＬＤＫ','４ＬＤＫ＋Ｓ','５ＤＫ',
     '５ＬＤＫ','５ＬＤＫ＋Ｓ','６ＤＫ',
     '６ＬＤＫ','６ＬＤＫ＋Ｓ',
     '７ＬＤＫ']

    madori_replace = dict(zip(madori,list(range(len(madori)))))

    df["間取り_順序"] = df["間取り"].replace({"オープンフロア":np.NaN,"スタジオ":np.NaN,"メゾネット":np.NaN}).replace(madori_replace).fillna(-1)

    #建築年をFE
    df["建築年_昭和"] = df.loc[df["建築年"].str.contains("昭和").fillna(False),"建築年"].replace({"昭和(.*)年":r'\1'},regex=True)
    df["建築年_平成"] = df.loc[df["建築年"].str.contains("平成").fillna(False),"建築年"].replace({"平成(.*)年":r'\1'},regex=True)

    df["建築年_昭和"] = df["建築年_昭和"].astype(float)
    df["建築年_平成"] = df["建築年_平成"].astype(float)
    
    #間取りコピペ
    df['L'] = df['間取り'].map(lambda x: 1 if 'Ｌ' in str(x) else 0)
    df['D'] = df['間取り'].map(lambda x: 1 if 'Ｄ' in str(x) else 0)
    df['K'] = df['間取り'].map(lambda x: 1 if 'Ｋ' in str(x) else 0)
    df['S'] = df['間取り'].map(lambda x: 1 if 'Ｓ' in str(x) else 0)
    df['R'] = df['間取り'].map(lambda x: 1 if 'Ｒ' in str(x) else 0)
    df['Maisonette'] = df['間取り'].map(lambda x: 1 if 'メゾネット' in str(x) else 0)
    df['OpenFloor'] = df['間取り'].map(lambda x: 1 if 'オープンフロア' in str(x) else 0)
    df['Studio'] = df['間取り'].map(lambda x: 1 if 'スタジオ' in str(x) else 0)
    
    #土地の形状FE
    keijou = ["正方形","長方形","ほぼ正方形","ほぼ長方形","ほぼ整形","台形","ほぼ台形","不整形","袋地等"]
    ind = list(range(len(keijou)))
    dic = dict(zip(keijou,ind))
    df["土地の形状_rank"] = df["土地の形状"].replace(dic)
    
    #建築年を西暦に
    wareki_list = list(df["建築年"].values)
    wareki_seireki(wareki_list)
    df["建築年_西暦"] = pd.Series(wareki_list)

    df["建築年_西暦"] = df["建築年_西暦"].replace({"戦前":1930})
    
    #取引時点をFE
    df["取引時点"] = df["取引時点"].replace({"(.*)年第(.*)":r'\1 \2',"(.*)四半期":r'\1'},regex=True)
    df = pd.concat([df,df["取引時点"].str.split(expand=True).rename(columns={0:"取引時点_年",1:"取引時点_四半期"})],axis=1)
    df = df.drop(["取引時点"],axis=1)
    df["取引時点_四半期"] = df["取引時点_四半期"].astype("int64")
    df["取引時点_年"] = df["取引時点_年"].astype("int64")
    
    #用途をFE
    master = pd.DataFrame(np.zeros((len(df),len(youto_master.columns))),columns=youto_master.columns)

    youto = df["用途"].str.split("、",expand=True)
    youtos = []
    for i in youto.columns:
        youtos.append(pd.get_dummies(youto[i]))

    for i in range(len(youtos)):
        b = youtos[i].copy()
        for col in master.columns:
            if (b.columns.values==col).sum()!=0:
                master[col] = master[col]+ b[col]

    df = pd.concat([df,master],axis=1)
    
    #延床面積をFE
    df["延床面積（㎡）"] = df["延床面積（㎡）"].replace({'2000㎡以上':2500,'10m^2未満':50}).astype(float)
    
    return df

train = FE(train)
test = FE(test)

print(train.shape,test.shape)
```

int →　そのまま（もしカテゴリなら指定）<br>
float →　そのまま<br>
floatだけど整数しかない　→　とりあえずそのまま（もしカテゴリにしたいなら指定）<br>
object →　基本的にはtrainもtestもlabelencodingしてからカテゴリに変更してる<br>

```python
def simple_preprocess(train,test,ID,TARGET,categories=None):
    
    #事前準備，ラベルエンコーダと，train,testを一緒にする(train,testの見分けは付くように)
    le = LabelEncoder()
    
    train["adversarial"] = 0
    test["adversarial"] = 1
    df = pd.concat([train,test],sort=False)
    
    objects = pd.DataFrame()
    col_obs = []
    col_fls = []
    col_ins = []
    col_fl_int = []
    
    #1列ずつ対処を変えていく
    print("object型のユニーク数：",end=' ')
    for col in df.drop([ID,TARGET],axis=1).columns:
        
        #型がobjectのものはユニーク数を出してから，labelencodeして数値に変更,かつcategoryに
        if df[col].dtype == object:
            
            col_obs.append(col)
            print(len(df[col].unique()),end=' ')
            df[col] = df[col].fillna("0")
            df[col] = le.fit_transform(df[col])
            df[col] = df[col].astype("category")
            
        #floatのものは，intなのかもしれないのでその見極めをする．純粋にfloatならそのまま
        elif df[col].dtype == float:
            
            if len(df) == df[col].fillna(0).apply(lambda x:x.is_integer()).sum():#すべてintegerならfloatintへ
                col_fl_int.append(col)
            else:
                col_fls.append(col)
        
        #intのものは，カテゴリなのか数値なのか怪しい
        elif train[col].dtype == "int64":
            col_ins.append(col)

    print("")
    print("----------")
    #処理の確認フェーズ
    print("shape:",train.shape)
    print("objects:",len(col_obs))
    display(df[col_obs].head())
    
    print("floats:",len(col_fls))
    display(df[col_fls].head())
    
    print("ints:",len(col_ins))
    display(df[col_ins].head())
    
    print("float to int",len(col_fl_int))
    display(df[col_fl_int].head())
    
    print("all:",len(col_obs)+len(col_fls)+len(col_ins)+len(col_fl_int))
    

    #int型とfloatint型はどんなものでどのくらいのユニーク数なのかは確認
    col_ins.extend(col_fl_int)
    print("int型とfloatだけどint型のユニーク数")
    for col in col_ins:
        print(col,"ユニーク数：",len(df[col].unique()))
    
    #指定されたカラムのみカテゴリカルに
    if categories != None:
        for col in categories:
            df[col] = df[col].astype("category")
    
    processed_train = df[df["adversarial"] == 0]
    processed_test = df[df["adversarial"] == 1]
    
    return processed_train,processed_test,col_ins
    
category = ["所在地コード","建築年_元号","間取り_順序","建築年_昭和","建築年_平成","土地の形状_rank","建築年_西暦","その他",
            "事務所","住宅","作業場","倉庫","共同住宅","工場","店舗","駐車場"]
processed_train,processed_test,col_ins = simple_preprocess(train,test,"id","y",categories=category)
```
